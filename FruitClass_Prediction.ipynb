Geometric Properties:

AREA: The area of the fruit.
PERIMETER: The perimeter or boundary length of the fruit.
MAJOR_AXIS: The length of the major axis of the fruit's shape.
MINOR_AXIS: The length of the minor axis of the fruit's shape.
ECCENTRICITY: A measure of the extent to which the shape deviates from being circular.
Shape Metrics:

EQDIASQ: Likely represents the square of the equivalent diameter of the fruit (a circular diameter that would have the same area).
SOLIDITY: The ratio of the area of the fruit to the area of its convex hull, indicating how solid or filled the shape is.
CONVEX_AREA: The area of the convex hull surrounding the fruit.
EXTENT: The ratio of the area of the fruit to the area of its bounding box.
ASPECT_RATIO: The ratio of the major axis to the minor axis, giving an idea of the shape’s elongation.
ROUNDNESS: A measure of how close the shape is to a perfect circle.
COMPACTNESS: The ratio of the square of the perimeter to the area, reflecting how compact the shape is.
SHAPEFACTOR_1 to SHAPEFACTOR_4: These are likely specific shape factors or metrics calculated to describe the fruit’s shape more comprehensively.

Statistical Features:

MeanRR, MeanRG, MeanRB: Mean values of different color channels (possibly Red, Green, and Blue).
StdDevRR, StdDevRG, StdDevRB: Standard deviations of the color channels, indicating color variability.
SkewRR, SkewRG, SkewRB: Skewness of the color channel distributions, describing asymmetry.
KurtosisRR, KurtosisRG, KurtosisRB: Kurtosis of the color channel distributions, indicating the "tailedness" or peakedness.
EntropyRR, EntropyRG, EntropyRB: Entropy measures of color channels, reflecting the amount of information or randomness in the color distribution.

Wavelet Transform Features:

ALLdaub4RR, ALLdaub4RG, ALLdaub4RB: These likely represent coefficients obtained from the Daubechies wavelet transform (order 4) for different color channels. Wavelet transforms are used for texture and pattern analysis.

```python
num_cols = len(df.drop(columns='Class').columns)
num_per_row = 3
num_rows = int(np.ceil(num_cols / num_per_row)) 
fig, axes = plt.subplots(num_rows, num_per_row, figsize=(5 * num_per_row, 5 * num_rows), sharey=True)
axes = axes.flatten()
for i,col in enumerate(df.drop(columns='Class').columns):
    sns.boxplot(x=df[col], ax=axes[i])
```

for col in df.drop(columns=['AREA', 'PERIMETER', 'MAJOR_AXIS', 'MINOR_AXIS','EQDIASQ','CONVEX_AREA','COMPACTNESS','MeanRR', 'MeanRG', 'MeanRB','StdDevRR','StdDevRB','SkewRG','ALLdaub4RR', 'ALLdaub4RG', 'ALLdaub4RB','Class']).columns:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    df=df[(df[col]>=lb)&(df[col]<=ub)]

```python
tdf=pd.get_dummies(df['Class'])
df=pd.concat([df,tdf],axis=1)
df.drop(columns='Class',axis=1,inplace=True)
print(list(df.columns))
```

```python
df[['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY']]=df[['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY']].astype(int)
print(df.info())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x=df.drop(columns=['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY'])
y=df[['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY']]
x_t,x_te,y_t,y_te=train_test_split(x,y,test_size=0.20,random_state=20)
ss=StandardScaler()
x_t=ss.fit_transform(x_t)
x_te=ss.transform(x_te)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dt=DecisionTreeClassifier()
path=dt.cost_complexity_pruning_path(x_t,y_t)
alphas=path.ccp_alphas
print(alphas)
```

```python
param={'ccp_alpha':alphas}
ndt=GridSearchCV(dt,param_grid=param,scoring='accuracy',cv=10)
ndt.fit(x_t,y_t)
print(ndt.best_params_)
print(ndt.best_score_)
bestalpha=ndt.best_params_['ccp_alpha']
```

```python
dt=DecisionTreeClassifier(ccp_alpha=bestalpha)
param={'criterion':['gini','entropy'],'max_depth':(list(np.arange(2,11))), 'min_samples_split':(list(np.arange(2,11))),
      'min_samples_leaf':(list(np.arange(2,11)))}
nde=GridSearchCV(dt,param_grid=param,scoring='accuracy',cv=10)
nde.fit(x_t,y_t)
print(nde.best_params_)
print(nde.best_score_)
nde=nde.best_estimator_
```

```python
pred=nde.predict(x_te)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
pred = nde.predict(x_te)
print("Classification Report:")
print(classification_report(y_te, pred, target_names=['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY'], zero_division=0))
print("Accuracy Score:")
print(accuracy_score(y_te, pred))
print("Confusion Matrix:")
print(confusion_matrix(y_te.values.argmax(axis=1), pred.argmax(axis=1)))
```

```python
names=['BERHI', 'DEGLET', 'DOKOL', 'IRAQI', 'ROTANA', 'SAFAVI', 'SOGAY']
plt.figure(figsize=(15,8))
sns.heatmap(confusion_matrix(y_te.values.argmax(axis=1), pred.argmax(axis=1)),annot=True ,xticklabels=names,
    yticklabels=names)
plt.show()
```
