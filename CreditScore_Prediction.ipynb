```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Imp\SamanyuK\Datasets\train.csv")
pd.set_option('display.max_columns',None)
df
```

```python
df.info()
```

```python
for col in df.columns[df.dtypes=='object']:
    print(df[col].value_counts(),'\n\n')
```

```python
col=['ID','Name','Type_of_Loan']
df.drop(columns=col,axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Occupation'] = le.fit_transform(df['Occupation'])
df['Payment_Behaviour']=le.fit_transform(df['Payment_Behaviour'])
df['Credit_Mix']=df['Credit_Mix'].apply(lambda x:0 if x=='Bad' else 2 if x=='Good' else 1)
df['Payment_of_Min_Amount']=df['Payment_of_Min_Amount'].apply(lambda x: 0 if x=='No' else 2 if x=='Yes' else 1)
df['Credit_Score']=df['Credit_Score'].apply(lambda x: 0 if x=='Poor' else 1 if x=='Standard' else 2)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
```

```python
x=df.drop(columns='Credit_Score')
y=df['Credit_Score']
x_t,x_te,y_t,y_te=train_test_split(x,y,test_size=0.25,random_state=20)
```

Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
path=dt.cost_complexity_pruning_path(x_t,y_t)
alphas=path.ccp_alphas
```

```python
param={'ccp_alpha':alphas}
ndt=RandomizedSearchCV(dt,param_distributions=param,cv=10,n_jobs=-1,random_state=20,scoring='accuracy',n_iter=100)
ndt.fit(x_t,y_t)
print(ndt.best_score_)
print(ndt.best_params_)
ba=ndt.best_params_['ccp_alpha']
```

```python
dt=DecisionTreeClassifier(ccp_alpha=ba)
param={'criterion':['gini','entropy'],'min_samples_split':list(np.arange(2,201)),'min_samples_leaf':list(np.arange(1,201)),
       'max_features':['sqrt','log2',None],'max_depth':list(np.arange(1,201))}
ndt=RandomizedSearchCV(dt,param_distributions=param,cv=10,n_jobs=-1,random_state=20,scoring='accuracy',n_iter=100)
ndt.fit(x_t,y_t)
print(ndt.best_score_)
print(ndt.best_params_)
```

```python
#ndt=ndt.best_estimator_
```

```python
pred=ndt.predict(x_te)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
cr1=classification_report(y_te,pred)
a1=accuracy_score(y_te,pred)
cm1=confusion_matrix(y_te,pred)
print(a1)
```

Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

rf = RandomForestClassifier()
param = {
    'n_estimators': [50],
    'criterion': ['gini', 'entropy'],
    'max_features': ['sqrt', 'log2', None]
}
nrf = RandomizedSearchCV(
    rf, 
    param_distributions=param, 
    n_jobs=-1, 
    n_iter=6,               
    cv=5,                   
    scoring='accuracy')

nrf.fit(x_t, y_t)

print(nrf.best_score_)
print(nrf.best_params_)
bs=nrf.best_score_
nrf = nrf.best_estimator_
```

```python
pred=nrf.predict(x_te)
```

```python
cr2=classification_report(y_te,pred)
a2=accuracy_score(y_te,pred)
cm2=confusion_matrix(y_te,pred)
print(a2)
```

```python
print("Decision Tree Classifier Training Accuracy and Testing Accuracy: ",ndt.best_score_," ",a1,"\n") 
print("Random Forest Classifier Training Accuracy and Testing Accuracy: ",nrf.best_score_," ",a2,"\n") 
```

```python
print("Decision Tree Classifier\n\n",cr1,"\n\n\n\nRandom Forest Classifier\n\n",cr2)
```
